# 2020

```python
# Leemos nuestro datset, utilizaremos los datos correspondientes a 2020
gastos_2020 =pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name="2020", header= 5)
gastos_2020
```

```python
# Seleccionamos las columnas de interés
gastos20_c = gastos_2020[['FECHA', "FOLIO",'PROVEEDOR', "DESCRIPCION", "TIPO GASTO", "IMPORTE", "IVA", "TOTAL MX"]]
gastos20_c.head(3)
```

```python
# Comprobamos si es necesario realizar alguna modificación de tipo de dato 
gastos20_c.info()
# Todo parece correcto
```

### Identificación de nulos 

```python
# Identificamos la existencia de nulos en nuestro dataframe
gastos20_c.isnull().sum()
```

### Columna FOLIO

```python
# Se identificaron 189 datos nulos en la columna FOLIO 
# Utilizaremos el siguiente formato : "XXXX" para sustituirlo y ser capaces de identificarlos posteriormente
gastos20_c["FOLIO"] = gastos20_c["FOLIO"].fillna("XXXX")
```

### Columna IMPORTE

```python
# Se identificaron 34 datos nulos en la columna IMPORTE
# Debido a que se desconoce el porque no se registró esa información, se sustituiran por el número 0
gastos20_c["IMPORTE"] = gastos20_c["IMPORTE"].fillna(0)
```

### Columna IVA

```python
# Se identificaron 268 datos nulos en la columna IVA
# Debido a que se desconoce el porque no se registró esa información, se sustituiran por el número 0
gastos20_c["IVA"] = gastos20_c["IVA"].fillna(0)
```

```python
# Comprobamos la imputación exitosa
gastos20_c.isnull().sum()
```
